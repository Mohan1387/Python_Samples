```python
# pandas and numpy for data manipulation
import pandas as pd
import numpy as np

# matplotlit and seaborn for visualizations
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 22
import seaborn as sns

# Suppress warnings from pandas
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
app_train = pd.read_csv('D:/CD/CD_variables_26_06_onehot.csv').reset_index(drop = True)
print('Training data shape: ', app_train.shape)
app_train.head()
```

```python
def plot_correlation_map( df ):
    corr = df.corr()
    sns_plot , ax = plt.subplots( figsize =( 50 , 50 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    sns_plot = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
    )
    
```

```python
plot_correlation_map(app_train)
```

```python
categories = cd.select_dtypes('object')
categories.columns
```

```python
categories.drop(['AGMTNO','PROPNO', 'PRODUCT_CODE', 'CITY'],axis=1,inplace= True)
```

```python
categories['IS_MANDATE_REGISTERED'] = cd['IS_MANDATE_REGISTERED']
```

```python
plt.figure(figsize = (15, 10))


sns.kdeplot(cd.loc[cd['IS_MANDATE_REGISTERED'] == 0, 'AGE'], label = 'IS_MANDATE_REGISTERED == 0')#,shade=True)

sns.kdeplot(cd.loc[cd['IS_MANDATE_REGISTERED'] == 1, 'AGE'], label = 'IS_MANDATE_REGISTERED == 1')#,shade=True)

# Labeling of plot
plt.xlabel('Age'); plt.ylabel('IS_MANDATE_REGISTERED'); plt.title('Distribution of Ages');
```

```python
plt.figure(figsize=(15,7))
for i in categories.columns:
#sns.pairplot(cd);
    if i != 'IS_MANDATE_REGISTERED':
        df = categories[[i,'IS_MANDATE_REGISTERED']] \
        .groupby([i,'IS_MANDATE_REGISTERED']) \
        .size()
        df = pd.DataFrame(df).reset_index()
        df.columns = [i,'IS_MANDATE_REGISTERED', 'Counts']
        df = df.pivot(i,'IS_MANDATE_REGISTERED', 'Counts').reset_index()

        df.plot.bar(figsize=(10,5)).set_xlabel(i)
        print('------------------- '+i+' ------------------')
        plt.savefig('D:/CD/Plot/'+i+'.png')

        plt.show()
```
