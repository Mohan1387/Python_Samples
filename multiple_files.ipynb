```python
import os
import pandas as pd
```

```python
path = "D:\CD\IFSC_CODE"
files = os.listdir(path)
files
```

```python
df = pd.DataFrame()
```

```python
for f in files:
    data = pd.read_excel(path+'\\'+f, header=None, skiprows=[0])
    data.columns = ['BANK','IFSC','MICR_CODE','BRANCH','ADDRESS','CONTACT','CITY','DISTRICT','STATE']
    print(f,data.columns)
    df = df.append(data)
```

```python
df
```

```python
df.to_csv(path+'\\'+"All_Bank_IFSC_CODE.csv", index=False)
```
