```python
query ="SELECT tb1.AGREEMENTNUMBER, \
tb1.TELEPHONE, \
tb2.ENQUIRINGMEMBERSHORTNAME, \
tb2.ENQUIRYPURPOSE, \
tb2.DATEOFENQUIRY \
FROM \
(SELECT * FROM BIUSR.TBL_TRN_CIBIL_ID_DETAILS) tb1 join \
(SELECT * FROM BIUSR.TBL_TRN_CIBIL_IQ_DETAILS \
WHERE ENQUIRINGMEMBERSHORTNAME = 'TVS CREDIT') tb2 \
ON tb1.AGREEMENTNUMBER = tb2.AGREEMENTNUMBER"
```

```python
Data = pd.read_sql(query, con=connection, chunksize=40000)
```
